# Challenge 3:3:7

```python
pop1 = np.random.binomial(10, 0.2, 10000)
pop2 = np.random.binomial(10,0.5, 10000)

sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

means1 = np.mean(sample1)
means2 = np.mean(sample2)

smp1std = sample1.std()
smp2std = sample2.std()

plt.hist(pop1, alpha=0.5, label='Population 1') 
plt.hist(pop2, alpha=0.5, label='Population 2') 
plt.legend(loc='upper right') 

plt.show()
```

```python
print(means1, means2, smp1std, smp2std) # stats from 100unit sample
```

1) I think that as the sample size increase the std should be reduced. I think this will happen because it will cancle out any major outliers.  When its reduced to 20units I think the data will vary greatly. Also, I think it will be very unreliable to make predictions about the whole population with.

```python
sample1 = np.random.choice(pop1, 1000, replace=True)
sample2 = np.random.choice(pop2, 1000, replace=True)

means1 = np.mean(sample1)
means2 = np.mean(sample2)

smp1std = sample1.std()
smp2std = sample2.std()

print(means1, means2, smp1std, smp2std)
```

with the sample size being increased to 1000, the std actually increased for both charts. This is the opposite of what I thought would happen.

```python
sample1 = np.random.choice(pop1, 20, replace=True)
sample2 = np.random.choice(pop2, 20, replace=True)

means1 = np.mean(sample1)
means2 = np.mean(sample2)

smp1std = sample1.std()
smp2std = sample2.std()

print(means1, means2, smp1std, smp2std)
```

After change the sample size to 20units, the std increased by like I thought it would. However, the change in the std was not nearly as drastic as i thought it would be.
