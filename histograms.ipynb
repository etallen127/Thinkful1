```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
normal = np.random.normal(50, 300, 100)
plt.hist(normal)
plt.axvline(normal.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(normal.mean() + normal.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(normal.mean()-normal.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```
