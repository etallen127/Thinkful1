```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
wine = pd.read_csv("winemag-data_first150k.csv")
```

```python
plt.figure(figsize=(16,12))


plt.subplot(4,3,1)
plt.scatter(x = wine['points'],y = wine['price'],marker= '.', s = 9)
plt.ylabel("Price")
plt.xlabel("Point Score")
plt.title("Price Vs. Wine Score")


plt.subplot(4,3,3)
histpoint = plt.hist(wine['points'], color ='lime', bins=np.arange(80,100), alpha=.6)
plt.title("Number Of Wines With The Same Point Rating")
plt.ylabel("Number of Wines With Scores")
plt.xlabel("Score Out of 100")


plt.subplot(4,3,2)
plt.scatter(x = wine['points'],y = wine['price'],marker= '.', s = 15)
plt.ylabel("price")
plt.xlabel("Point Score")
plt.title("Price Vs. Wine Score Above $500")
plt.ylim([350,2400])

plt.tight_layout()
plt.show()
```

```python
""" 
    In this data set, the first and second graphs show the points earned vs. the price of the wine. 
While the third chart shows the number of wines at each ratings.  From the first and second chart, 
we can see that while increasing price does improve the chances of a high rating,
there are still plenty of cheaper wine that are able to make the cut.  With the visualization of the third graph
and wine['price'].mean()' equaling 33.13. I think that price have very little impact on the quality/point scores 
of the wine tested.

"""
```

```python
abc = pd.read_csv("ted_main.csv")
ted = abc[['title','name','comments','views','languages']]
ted['comment percent'] = ted["comments"] / ted["views"] * 100
```

```python
plt.figure(figsize=(12,12))

plt.subplot(1,2,1)
plt.scatter(x = ted['languages'],y = ted['views'],marker = '.', s = 10)
plt.ylabel("Views in the 10Millions")
plt.xlabel("Number of Languages")
plt.title("Ted Talk Views v Number Languages")
plt.ylim(1000,50000000)

plt.subplot(1,2,2)
plt.hist(x = ted['languages'], bins = 70, color='silver')
plt.ylabel("Ted Talks with X Languages")
plt.xlabel("Number of Translated Languages")
plt.title("Number of Videos for Each Number of Languages")

plt.tight_layout
plt.show()
```

```python
"""This scatter chart shows the relationship between the number of languages the broadcast was translated to 
and the total number of views the ted talked recieved. The second graph shows the number of ted talks based on
the number of languages they were translated into.  With these charts I want to find out what would lead to more views.
However, I was unale to make a clear prediction, because there was no way for me to tell if the presentation was
originally made in a large number of languages, helping it gain views, or if the video was translated into more languages 
because it received so many views.  Last, with the second chart a histogram, showing a large number of videos with 
20 to 35 languages. I would imagine most talks start with 20-35 and gain more as they gaim popularity and view.  However,
I would need more data to be sure"""
```

```python
plt.figure(figsize=(12,12))

plt.subplot(1,2,1)
plt.scatter(x = ted['languages'],y = ted['views'],marker = '.', s = 10)
plt.ylabel("Views in the 10Millions")
plt.xlabel("Number of Languages")
plt.title("Ted Talk Views v Number Languages")
plt.ylim(1000,50000000)

plt.subplot(1,2,2)
plt.hist(x = ted['languages'], bins = 70, color='silver')
plt.ylabel("Ted Talks with X Languages")
plt.xlabel("Number of Translated Languages")
plt.title("Number of Videos for Each Number of Languages")

plt.subplot(1,2,1)
plt.scatter(x = ted['languages'],y = ted['views'],color = 'purple',marker = 'o', s = 6)
plt.ylabel("Views")
plt.xlabel("Number of Languages")
plt.title("0-30 Languages")
plt.ylim(1000,2500000)
plt.xlim(0,30)

plt.subplot(1,2,2)
plt.scatter(x = ted['languages'],y = ted['views'],color = 'green',marker = 'o', s = 6)
plt.ylabel("Views in 10Millions")
plt.xlabel("Number of Languages")
plt.title("30-65 Languages")
plt.ylim(100000,15000000)
plt.xlim(0,65)

plt.tight_layout
plt.show()
```
